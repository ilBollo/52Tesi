```python
import re
from langchain.text_splitter import TextSplitter
from langchain.schema import Document

class RegexTextSplitter(TextSplitter):
    def __init__(self, pattern: str, keep_separator: bool = False):
        """
        Splitter personalizzato basato su regex.

        Args:
            pattern (str): Il pattern regex per suddividere il testo.
            keep_separator (bool): Se mantenere il separatore nel chunk.
        """
        self.pattern = pattern
        self.keep_separator = keep_separator

    def split_text(self, text: str) -> list[str]:
        matches = re.finditer(self.pattern, text, flags=re.DOTALL)
        chunks = []
        for match in matches:
            chunk = match.group(0)
            if not self.keep_separator:
                # Rimuove il separatore (es. "Domanda:") se richiesto
                chunk = chunk.lstrip("Domanda:")
            chunks.append(chunk.strip())
        return chunks


# Concatenare il testo di tutti i documenti in un unico testo
full_text = "\n".join([doc.page_content for doc in docs])

# Definire il pattern regex
pattern = r"(Domanda:.*?)(?=Domanda:|$)"  # Cattura tutto da 'Domanda:' fino alla prossima 'Domanda:' o alla fine

# Creare e usare il TextSplitter
text_splitter = RegexTextSplitter(pattern=pattern, keep_separator=True)
splits = text_splitter.split_text(full_text)


from langchain.schema import Document

# Verifica il tipo di ogni elemento per evitare duplicazioni
splits = [
    item if isinstance(item, Document) else Document(page_content=item)
    for item in splits
]

# Filtrare i chunk con meno di 200 caratteri
#filtered_splits = [chunk for chunk in splits if len(chunk) >= 200]

# Stampare i chunk filtrati
for i, chunk in enumerate(splits):
    print(f"Chunk {i + 1}:\n{chunk}\n")
```

```python
from langchain_community.embeddings import OllamaEmbeddings

embeddings = OllamaEmbeddings(
    model = "nomic-embed-text"
)
```

```python
import os
from langchain.vectorstores import FAISS
from langchain.embeddings import OpenAIEmbeddings  # O il modello che stai usando
from langchain.schema import Document

# Supponiamo che `filtered_splits` sia un array di stringhe
# Converte ogni stringa in un oggetto Document

# Controlla se l'indice FAISS esiste già
if os.path.exists("domande"):
    # Carica l'indice FAISS dalla directory esistente
    faiss_index = FAISS.load_local(
        "domande",
        embeddings,
        allow_dangerous_deserialization=True
    )
else:
    # Crea un nuovo indice FAISS
    faiss_index = FAISS.from_documents(
        splits,  # Deve essere una lista di Document
        embeddings        # Modello di embedding
    )

    # Salva l'indice nella directory "domande"
    faiss_index.save_local("domande")
```

```python
#Create Retriever
retriever = faiss_index.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 3}
)
```

```python
#Create Prompt
from langchain import hub
prompt = hub.pull("rlm/rag-prompt")
```

```python
from langchain_core.runnables import RunnablePassthrough

def format_documents(docs):
    return "\n\n".join(doc.page_content for doc in docs)

rag_chain = (
    {
    "context": retriever | format_documents,
    "question": RunnablePassthrough()
} 
| prompt 
| model
| StrOutputParser()
)
```

```python
def query(query):
    answer = rag_chain.invoke(query)
    return answer

def query_stream(query):
    for chunk in rag_chain.stream(query):
        print(  chunk, end="", flush=True)
```
